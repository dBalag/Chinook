### 3.Visualize the Data
Create some visualizations to explore the data further.

This project uses Seaborn, a Python data visualization library, to create informative and visual statistical graphics. Matplotlib, a comprehensive plotting library, is used to customize and fine-tune the visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of total invoice amounts
plt.figure(figsize=(10, 6))
sns.histplot(invoice_cleaned['total'], kde=True, bins=30)
plt.title('Distribution of Total Invoice Amounts')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

Around 2$ is the Total Amount of more invoices (almost 120)

```python
# TOP 10 GENRES X NR OF TRACKS

# Merge track and genre data to get genre names for each track
track_genre = pd.merge(track_cleaned, genre, on='genre_id')

# Get the top 10 most popular genres
top_genres = track_genre['name_y'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title('Top 10 Most Popular Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.show()
```

```python
# MONTHLY SALES TREND

# Convert invoice_date to datetime
invoice_cleaned['invoice_date'] = pd.to_datetime(invoice_cleaned['invoice_date'])

# Group by month and sum the total amount
monthly_sales = invoice_cleaned.resample('M', on='invoice_date').sum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales['total'])
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# DISTRIBUTION OF TRACKS X MEDIA TYPE

# Merge track and media_type data to get media type names for each track
track_media = pd.merge(track_cleaned, media_type, on='media_type_id')

# Plot the distribution of tracks by media type
media_type_counts = track_media['name_y'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=media_type_counts.index, y=media_type_counts.values)
plt.title('Distribution of Tracks by Media Type')
plt.xlabel('Media Type')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.show()
```
