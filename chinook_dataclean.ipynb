```python
# Function for basic data cleaning
def clean_data(df, df_name):
    print(f"Cleaning {df_name}...")
    display(df.head())
    display(df.info())
    missing_values = df.isnull().sum()
    print(f"Missing values in {df_name}:\n{missing_values[missing_values > 0]}")
    df.drop_duplicates(inplace=True)
    df.dropna(inplace=True)
    display(df.info())
    return df
```

```python
# Clean each dataset
media_type = clean_data(media_type, 'media_type')
playlist = clean_data(playlist, 'playlist')
playlist_track = clean_data(playlist_track, 'playlist_track')
album = clean_data(album, 'album')
artist = clean_data(artist, 'artist')
employee = clean_data(employee, 'employee')
invoice = clean_data(invoice, 'invoice')
invoice_line = clean_data(invoice_line, 'invoice_line')
track = clean_data(track, 'track')
genre = clean_data(genre, 'genre')
customer = clean_data(customer, 'customer')
```

```python
# Convert date columns to datetime
def convert_to_datetime(df, column_name):
    df[column_name] = pd.to_datetime(df[column_name])
    return df

invoice = convert_to_datetime(invoice, 'invoice_date')
employee = convert_to_datetime(employee, 'birth_date')
employee = convert_to_datetime(employee, 'hire_date')
```

```python
# Convert columns to numeric where appropriate
def convert_to_numeric(df, column_name):
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return df

track = convert_to_numeric(track, 'milliseconds')
track = convert_to_numeric(track, 'bytes')
track = convert_to_numeric(track, 'unit_price')
```
