```python
#Load Datasets
import pandas as pd
customer = pd.read_csv('customer.csv')
employee = pd.read_csv('employee.csv')
invoice = pd.read_csv('invoice.csv')
track = pd.read_csv('track.csv')
```

```python
#Fill missing values Customer Dataset
customer['company'].fillna('N/A', inplace=True)
customer['state'].fillna('Unknown', inplace=True)
customer['postal_code'].fillna('Unknown', inplace=True)
customer['phone'].fillna('Unknown', inplace=True)
customer['fax'].fillna('N/A', inplace=True)
```

```python
#Fill missing values Employee Dataset
employee['reports_to'].fillna(0, inplace=True)
```

```python
#Fill missing values Invoice Dataset
invoice['billing_state'].fillna('Unknown', inplace=True)
invoice['billing_postal_code'].fillna('Unknown', inplace=True)
```

```python
#Fill missing values Track Dataset
track['composer'].fillna('Unknown', inplace=True)
```

```python
#Review and ensure changes
print(customer.isnull().sum())
print(employee.isnull().sum())
print(invoice.isnull().sum())
print(track.isnull().sum())
```

```python
#save cleaned datasets
customer.to_csv('customer_cleaned.csv', index=False)
employee.to_csv('employee_cleaned.csv', index=False)
invoice.to_csv('invoice_cleaned.csv', index=False)
track.to_csv('track_cleaned.csv', index=False)
```

```python
#Unnamed column: When loading this CSV back into a DataFrame, the index column becomes an unnamed column because it doesnâ€™t have a header.
#To Check if 'Unnamed: 0' column exists before dropping
if 'Unnamed: 0' in customer.columns:
    customer = customer.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in employee.columns:
    employee = employee.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in invoice.columns:
    invoice = invoice.drop(columns=['Unnamed: 0'])
if 'Unnamed: 0' in track.columns:
    track = track.drop(columns=['Unnamed: 0'])
```

```python
#Review and ensure changes, Unnamed column removed
print(customer.isnull().sum())
print(employee.isnull().sum())
print(invoice.isnull().sum())
print(track.isnull().sum())
```
