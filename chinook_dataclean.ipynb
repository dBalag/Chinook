```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load data
media_type = pd.read_csv('media_type.csv')
playlist = pd.read_csv('playlist.csv')
playlist_track = pd.read_csv('playlist_track.csv')
album = pd.read_csv('album.csv')
artist = pd.read_csv('artist.csv')
employee = pd.read_csv('employee.csv')
invoice = pd.read_csv('invoice.csv')
invoice_line = pd.read_csv('invoice_line.csv')
track = pd.read_csv('track.csv')
genre = pd.read_csv('genre.csv')
customer = pd.read_csv('customer.csv')

# Handle missing values
invoice.fillna(0, inplace=True)
invoice_line.dropna(subset=['track_id'], inplace=True)

# Remove duplicates
playlist.drop_duplicates(inplace=True)
playlist_track.drop_duplicates(inplace=True)
album.drop_duplicates(inplace=True)
artist.drop_duplicates(inplace=True)
employee.drop_duplicates(inplace=True)
media_type.drop_duplicates(inplace=True)
invoice.drop_duplicates(inplace=True)
invoice_line.drop_duplicates(inplace=True)
track.drop_duplicates(inplace=True)
genre.drop_duplicates(inplace=True)
customer.drop_duplicates(inplace=True)

# Fix data types
invoice['invoice_date'] = pd.to_datetime(invoice['invoice_date'])
invoice['total'] = invoice['total'].astype(float)
track['milliseconds'] = track['milliseconds'].astype(int)
track['bytes'] = track['bytes'].astype(int)

# Handle outliers
Q1 = invoice['total'].quantile(0.25)
Q3 = invoice['total'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
invoice = invoice[(invoice['total'] >= lower_bound) & (invoice['total'] <= upper_bound)]

# Normalize data
scaler = MinMaxScaler()
invoice['total'] = scaler.fit_transform(invoice[['total']])

# Merge DataFrames
invoice_customer = pd.merge(invoice, customer, on='customer_id')
invoice_data = pd.merge(invoice_customer, invoice_line, on='invoice_id')


# Drop 'Unnamed: 0' columns from both DataFrames before merging
invoice_data = invoice_data.drop('Unnamed: 0_x', axis=1)
track = track.drop('Unnamed: 0', axis=1)

# Now perform the merge
invoice_data = pd.merge(invoice_data, track, on='track_id')
invoice_data = pd.merge(invoice_data, genre, on='genre_id')


# Display the cleaned and merged DataFrame
print(invoice_data.head())
print(invoice_data.info())
```
