```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the genre.csv file (it has 'name' and 'revenues' columns)
genre_data = pd.read_csv('rev x genre.csv')

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(genre_data['name'], genre_data['revenues'])
plt.xlabel('Name')
plt.ylabel('Revenues')
plt.title('Revenue by Genre')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('revs x genre x location .csv')

# Check the column names in DataFrame, as doesn't match with SQL defined.
print(data.columns)

```

```python
# Pivot the data to create a matrix for the heatmap
heatmap_data = data.pivot_table(index='genre', columns='billing_city', values='total', aggfunc='sum')

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Revenue by Genre and Billing City')
plt.xlabel('Billing City')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
